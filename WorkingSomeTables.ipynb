```python
import pandas as pd
```

```python
from IPython.display import Image
```

```python
calendar = pd.read_csv('boston-airbnb-open-data/calendar_sept.csv', index_col=0)
reviews = pd.read_csv('boston-airbnb-open-data/reviews_sample.csv', index_col=0)
listings = pd.read_csv('boston-airbnb-open-data/listings.csv', index_col=0)
```

```python
# Бронирования квартиры
calendar.head(2)
```

```python
# Отзывы о квартирах
reviews.head(2)
```

```python
# Информация о квартирах
listings.head(2)
```

```python
# В pandas существует два типа объединения таблиц: merge and join
# merge связывает таблицы по каким-то признакам
# join связывает таблицы по какому-то индексу
```

```python
# Добавим к квартире отзывы людей (merge): left_on, right_on - названия столбцов для объединения
pd.merge(listings, reviews, left_on=['id'], 
         right_on=['listing_id'], how='right', indicator=True).head(1)
```

```python
# Рассмотрим метод join
calendar.set_index('listing_id', inplace=True)
```

```python
reviews.set_index('listing_id', inplace=True)
```

```python
calendar.join(reviews, lsuffix='listing_id', rsuffix='listing_id')
```
